```python
train_df = pd.DataFrame(columns=["id", "text"])
#temp = pd.DataFrame(columns=["id", "text"])
for i in tqdm(range(0,len(id_list),100)):
    tweets = api.statuses_lookup(id_list[i:i+100]) # id_list is the list of tweet ids
    for i in tweets:
        temp = pd.DataFrame({"id":i.id,"text":i.text}, index=[0])
        #print(temp)
        train_df = train_df.append(temp)
        
train_df
```
